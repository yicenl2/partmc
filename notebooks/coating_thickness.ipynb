# Coating Thickness (rad_part - rad_core)

```python
directory = 'sce_completed/scenarios/scenario_0074'
diam_coating_dist = np.zeros((2,100,25,50,180))

for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ('out/urban_plume_00000' in filename and '.nc' in filename):
            t_index = int(filename[-13:-11]) - 1
            sce_index = int(filename[-38:-36])
            ncf = scipy.io.netcdf_file(filename)
            diam_edges = ncf.variables["diam_edges"].data *1e6
            coating_edges = ncf.variables["coating_edges"].data * 1e9
            diam = ncf.variables["diam"].data * 1e6 # m to µm
            diam_coating_dist[0,sce_index, t_index,:,:] = ncf.variables["coating_dist_pr"].data * 1e9 # m to nm
            diam_coating_dist[1,sce_index, t_index,:,:] = ncf.variables["coating_dist_comp"].data * 1e9

plt.figure(figsize=(12,4))
X, Y = np.meshgrid(diam_edges, coating_edges)
case_index = 74
t_index = 4

plt.subplot(1,2,1)
p = plt.pcolormesh (X, Y, diam_coating_dist[0, sce_index, t_index,:,:], 
                    norm = mpl.colors.LogNorm(vmin=1e0, vmax=1e13), linewidths = 0.1)
plt.xlim(1e-2,1e3)
plt.xscale('log')

plt.ylim(5e-4,5e3)
plt.ylabel('coating_dist_pr')
plt.yscale('log')
plt.text(0.015, 1.1e3, r'sce:%d T:%d h'% (sce_index, t_index))
plt.grid(True)
#*************
plt.subplot(1,2,2)
p = plt.pcolormesh (X, Y, diam_coating_dist[1, sce_index, t_index,:,:], 
                    norm = mpl.colors.LogNorm(vmin=1e0, vmax=1e13), linewidths = 0.1)
plt.xlim(1e-2,1e3)
plt.xscale('log')

plt.ylim(5e-4,5e3)
plt.ylabel('coating_dist_comp')
plt.yscale('log')
plt.text(0.015, 1.1e3, r'sce:%d T:%d h'% (sce_index, t_index))
plt.grid(True)
# plt.savefig('/data/keeling/a/yicenl2/d/images/' + s[i] + '_On.png')
```
