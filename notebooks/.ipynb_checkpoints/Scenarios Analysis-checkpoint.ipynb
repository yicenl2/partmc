```python
directory = 'sce_completed/scenarios'
gamma_pop = np.zeros((2,100,25))
bulk_spec_mass = np.zeros((13,100,25)) #tot_dry, tot_so4, tot_no3
chi = np.zeros((2,100,25))

#Bulk properties
for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ("urban_plume_process.nc" in filename):
            #print(filename[-30:-27])
            sce_ind      =  int(filename[-30:-27])
            ncf  = scipy.io.netcdf_file(filename, mmap=False)
            gamma_pop[0, sce_ind, :] = ncf.variables["gamma_pop_pr"].data
            gamma_pop[1, sce_ind, :] = ncf.variables["gamma_pop_comp"].data
            chi[0,sce_ind, :] = ncf.variables["chi"].data
            bulk_spec_mass[0, sce_ind, :] = ncf.variables["tot_dry_mass_conc"].data*1e9
            bulk_spec_mass[1, sce_ind, :] = ncf.variables["tot_no3_mass_conc"].data*1e9
            bulk_spec_mass[2, sce_ind, :] = ncf.variables["tot_so4_mass_conc"].data*1e9
            bulk_spec_mass[3, sce_ind, :] = ncf.variables["tot_oin_mass_conc"].data*1e9
            bulk_spec_mass[4, sce_ind, :] = ncf.variables["tot_cl_mass_conc"].data*1e9
            bulk_spec_mass[5, sce_ind, :] = ncf.variables["tot_na_mass_conc"].data*1e9
            bulk_spec_mass[6, sce_ind, :] = ncf.variables["tot_h2o_mass_conc"].data*1e9
            bulk_spec_mass[7, sce_ind, :] = ncf.variables["tot_nh4_mass_conc"].data*1e9
            bulk_spec_mass[8, sce_ind, :] = ncf.variables["tot_bc_mass_conc"].data*1e9
            bulk_spec_mass[9, sce_ind, :] = ncf.variables["tot_ca_mass_conc"].data*1e9
            bulk_spec_mass[10, sce_ind, :] = ncf.variables["tot_co3_mass_conc"].data*1e9
            bulk_spec_mass[11, sce_ind, :] = ncf.variables["tot_oc_mass_conc"].data*1e9
            bulk_spec_mass[12, sce_ind, :] = ncf.variables["tot_soa_mass_conc"].data*1e9
      
gamma_pop_error = (gamma_pop[1,:,:].ravel() - 
                   gamma_pop[0,:,:].ravel())/gamma_pop[0,:,:].ravel()
d = {"gamma_pop_error": gamma_pop_error * 100, # convert to percentage
     "chi": chi[0,:,:].ravel(),
     "no3_frac": bulk_spec_mass[1,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "so4_frac": bulk_spec_mass[2,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "oin_frac": bulk_spec_mass[3,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "cl_frac": bulk_spec_mass[4,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "na_frac": bulk_spec_mass[5,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "h2o_frac": bulk_spec_mass[6,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "nh4_frac": bulk_spec_mass[7,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "bc_frac": bulk_spec_mass[8,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "ca_frac": bulk_spec_mass[9,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "co3_frac": bulk_spec_mass[10,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "oc_frac": bulk_spec_mass[11,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "soa_frac": bulk_spec_mass[12,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "wi": bulk_spec_mass[2, :, :].ravel()/(bulk_spec_mass[1, :, :].ravel() 
                                          + bulk_spec_mass[2, :, :].ravel())}
df = pd.DataFrame(data=d)
```
