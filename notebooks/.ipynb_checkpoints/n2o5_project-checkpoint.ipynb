## $\gamma_{\rm pop, ca}$ vs. $\gamma_{\rm pop, pr}$

```python
# Read data
directory = 'sce_completed/scenarios'
gamma_pop = np.zeros((2,100,25))
chi = np.zeros((100,25))

for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ("out/urban_plume_process.nc" in filename):
            sce_ind      =  int(filename[-30:-27])
            ncf  = scipy.io.netcdf_file(filename, mmap=False)
            gamma_pop[0, sce_ind, :] = ncf.variables["gamma_pop_pr"].data
            gamma_pop[1, sce_ind, :] = ncf.variables["gamma_pop_comp"].data
            time = ncf.variables["time"].data/3600
            chi[sce_ind, :] = ncf.variables["chi"].data

gamma_pop_error = (gamma_pop[1,:,:].ravel() - 
                   gamma_pop[0,:,:].ravel())/gamma_pop[0,:,:].ravel()
# nighttime cases only          
d = {"gamma_pop_pr": gamma_pop[0,:,:].ravel(),
     "gamma_pop_comp": gamma_pop[1,:,:].ravel(),
     "gamma_pop_error": gamma_pop_error * 100,
     "chi": chi[:,:].ravel()}
df = pd.DataFrame(data=d)
case_index = df.index.values
sce_index  = (case_index/25).astype(int)
hour_index = case_index % 25
df_ngt = df[((df.index.values % 25) >= 12) & ((df.index.values % 25) < 24)]
print(df_ngt.sort_values(by='gamma_pop_error', ascending=False))

df_on = df_ngt[((df_ngt["gamma_pop_error"]) < 20)]
df_above = df_ngt[((df_ngt["gamma_pop_error"]) >= 20)]
df_below = df_ngt[((df_ngt["gamma_pop_error"]) <= -20)]
```

```python
# All cases (daytime & nighttime, sea salts & dusts)
plt.figure(figsize=(7,6))
plt.xlim(0,2.5e-2);plt.ylim(0,2.5e-2)
plt.plot([0,3e-2],[0,3e-2],'--', color='k')
a = plt.scatter(gamma_pop[0,:,:].ravel(), gamma_pop[1,:,:].ravel(), c=chi.ravel())#, s=30,facecolors='none', edgecolors='k', c=chi.ravel())
cbar = plt.colorbar(a)
plt.clim(0,1.0)
plt.xlabel(r'Particle-Resolved $\gamma_{\rm pop,PR}$')
plt.ylabel(r'Composition-Averaged $\gamma_{\rm pop,CA}$')
cbar.ax.set_ylabel('Mixing state $\chi$')
```

## $\gamma_{\rm pop, mix}$ vs. $\gamma_{\rm pop, pr}$

```python
# Read data
directory = 'sce_completed/scenarios'
gamma_pop = np.zeros((2,100,25))
chi = np.zeros((100,25))
bulk_spec_mass = np.zeros((4,100,25))

for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ("out/urban_plume_process.nc" in filename):
            sce_ind      =  int(filename[-30:-27])
            ncf  = scipy.io.netcdf_file(filename, mmap=False)
            gamma_pop[0, sce_ind, :] = ncf.variables["gamma_pop_pr"].data
            gamma_pop[1, sce_ind, :] = ncf.variables["gamma_pop_mix"].data
            time = ncf.variables["time"].data/3600
            bulk_spec_mass[0, sce_ind, :] = ncf.variables["tot_dry_mass_conc"].data*1e9
            bulk_spec_mass[1, sce_ind, :] = ncf.variables["tot_no3_mass_conc"].data*1e9
            bulk_spec_mass[2, sce_ind, :] = ncf.variables["bulk_poa_masses"].data*1e9
            bulk_spec_mass[3, sce_ind, :] = ncf.variables["bulk_soa_masses"].data*1e9
            #time = ncf.variables["time"].data/3600
            chi[sce_ind, :] = ncf.variables["chi"].data
            
d = {"bulk_no3_masses": bulk_spec_mass[1, :, :].ravel(), # convert to percentage
     "no3_frac": bulk_spec_mass[1,:,:].ravel()/bulk_spec_mass[0,:,:].ravel(),
     "no3/org": bulk_spec_mass[1,:,:].ravel()/(bulk_spec_mass[2,:,:].ravel()+bulk_spec_mass[3,:,:].ravel()),
     "org/no3":(bulk_spec_mass[2,:,:].ravel()+bulk_spec_mass[3,:,:].ravel())/bulk_spec_mass[1,:,:].ravel(),
     "org": bulk_spec_mass[2,:,:].ravel()+bulk_spec_mass[3,:,:].ravel()}
df = pd.DataFrame(data=d)
```

```python
# Make plot
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.xlim(0,2.5e-2);plt.ylim(0,2.5e-2)
plt.plot([0,3e-2],[0,3e-2],'--', color='k')
a = plt.scatter(gamma_pop[0,:,:].ravel(), gamma_pop[1,:,:].ravel(), c=chi.ravel())
cbar = plt.colorbar(a)
plt.clim(0,1.0)
plt.xlabel(r'Particle-Resolved $\gamma_{\rm pop,PR}$')
plt.ylabel(r'Well-Mixed $\gamma_{\rm pop,MIX}$')
cbar.ax.set_ylabel('Mixing state $\chi$')

plt.subplot(1,2,2)
plt.xlim(0,2.5e-2);plt.ylim(0,2.5e-2)
plt.plot([0,3e-2],[0,3e-2],'--', color='k')
a = plt.scatter(gamma_pop[0,:,:].ravel(), gamma_pop[1,:,:].ravel(), c=df["org/no3"])
cbar = plt.colorbar(a)
plt.clim(0,5)
plt.xlabel(r'Particle-resolved $\gamma_{\rm pop,PR}$')
plt.ylabel(r'Well-mixed $\gamma_{\rm pop,mix}$')
cbar.ax.set_ylabel(r'$ \frac{[\rm POA]+[\rm SOA]}{[\rm NO_3]}$')
plt.tight_layout()
```

```python
# Make plot
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.xlim(0,2.5e-2);plt.ylim(0,2.5e-2)
plt.plot([0,3e-2],[0,3e-2],'--', color='k')
a = plt.scatter(gamma_pop[0,:,:].ravel(), gamma_pop[1,:,:].ravel(), c=df["bulk_no3_masses"])
cbar = plt.colorbar(a)
plt.clim(0,13)
plt.xlabel(r'Particle-resolved $\gamma_{\rm N_2O_5}$')
plt.ylabel(r'Well-mixed $\gamma_{\rm N_2O_5}$')
cbar.ax.set_ylabel(r'$[\rm NO_3]\,\ (µg/m^3)$')
plt.tight_layout()

plt.subplot(1,2,2)
plt.xlim(0,2.5e-2);plt.ylim(0,2.5e-2)
plt.plot([0,3e-2],[0,3e-2],'--', color='k')
a = plt.scatter(gamma_pop[0,:,:].ravel(), gamma_pop[1,:,:].ravel(), c=df["org"])
cbar = plt.colorbar(a)
plt.clim(0,10)
plt.xlabel(r'Particle-resolved $\gamma_{\rm N_2O_5}$')
plt.ylabel(r'Well-mixed $\gamma_{\rm N_2O_5}$')
cbar.ax.set_ylabel(r'$[\rm POA]+[\rm SOA] \, (\mu g /m^3)$')
plt.tight_layout()
```

# Size resolved gamma (hgamma)

1. For PR:
<br>hgamma = h1/h2
<br>h1 = surf_area * gamma
<br>h2 = surf_area
2. For CA:
<br>mgamma = m1/m2
<br>m1 = mass * gamma
<br>m2 = mass

```python
# Read data
directory = 'sce_completed/scenarios'
gamma_pr = np.zeros((100,25,180)) # for PR
gamma_comp = np.zeros((100,25,180)) # for CA
surf_area_dist = np.zeros((100,25,180)) # for PR
mass_dist = np.zeros((100,25,180)) # for CA
hgamma = np.zeros((100,25,180)) # for PR
mgamma = np.zeros((100,25,180)) # for CA

for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ('out/urban_plume_00000' in filename and '.nc' in filename):
            t_index = int(filename[-13:-11]) - 1
            sce_index = int(filename[-38:-36])
            ncf = scipy.io.netcdf_file(filename)
            diam = ncf.variables["diam"].data* 1e6 # m to µm
            gamma_pr[sce_index,t_index,:] = ncf.variables["gamma_surf_pr"].data*1e6 # m3 to cm-3
            gamma_comp[sce_index,t_index,:] = ncf.variables["gamma_mass_comp"].data*1e9 # kg to µg
            surf_area_dist[sce_index,t_index,:] = ncf.variables["surf_area_dist_pr"].data*1e6 # m3 to cm-3
            mass_dist[sce_index,t_index,:] = ncf.variables["mass_dist_comp"].data*1e9 # kg to µg
            hgamma[sce_index,t_index,:] = gamma_pr[sce_index, t_index,:]/surf_area_dist[sce_index,t_index,:]
            mgamma[sce_index,t_index,:] = gamma_comp[sce_index, t_index,:]/mass_dist[sce_index,t_index,:]
```

```python
# Make plot
plt.figure(figsize=(18,4))

# Above 1:1 Line (59,14h)
plt.subplot(1,3,1)
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,0.025)
plt.plot(diam, hgamma[59,14,:], label='PR')
plt.plot(diam, mgamma[59,14,:], label='CA')
plt.ylabel('Size-Resolved Gamma')
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 0.022, r'sce:%d T:%d h'% (59,14))
plt.grid()

# On 1:1 Line (82,17h)
plt.subplot(1,3,2)
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,0.025)
plt.plot(diam, hgamma[82,17,:], label='PR')
plt.plot(diam, mgamma[82,17,:], label='CA')
plt.ylabel('Size-Resolved Gamma')
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 0.022, r'sce:%d T:%d h'% (82,17))
plt.grid()

# Below 1:1 Line (47,18h)
plt.subplot(1,3,3)
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,0.025)
plt.plot(diam, hgamma[47,18,:], label='PR')
plt.plot(diam, mgamma[47,18,:], label='CA')
plt.ylabel('Size-Resolved Gamma')
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 0.022, r'sce:%d T:%d h'% (47,18))
plt.grid()

plt.tight_layout()
# plt.savefig('/data/keeling/a/yicenl2/d/images/h2_hgamma/hgamma.png')
```
