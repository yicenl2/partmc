# 1D Size Distribution (h1, surf_area_concs * gamma_part)

```python
directory = 'sce_completed/scenarios'
gamma_1d_dist = np.zeros((2,100,25,180))

for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ('out/urban_plume_00000' in filename and '.nc' in filename):
            t_index = int(filename[-13:-11]) - 1
            sce_index = int(filename[-38:-36])
            ncf = scipy.io.netcdf_file(filename)
            diam = ncf.variables["diam"].data* 1e6 # m to Âµm
            gamma_1d_dist[0,sce_index, t_index,:] = ncf.variables["gamma_size_pr"].data*1e6 # m3 to cm-3
            gamma_1d_dist[1,sce_index, t_index,:] = ncf.variables["gamma_size_comp"].data*1e6
```

```python
plt.figure(figsize=(15,4))

# Above 1:1 Line (74,4h)
plt.subplot(1,3,1)
sce_index = 74
t_index = 4
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,32)
plt.plot(diam, gamma_1d_dist[0, sce_index, t_index, :], label='PR')
plt.plot(diam, gamma_1d_dist[1, sce_index, t_index, :], label='CA', ls='--')
plt.ylabel(r'$h1$')
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 29, r'sce:%d T:%d h'% (sce_index, t_index))
plt.legend(loc=1)
plt.grid()

# On 1:1 Line (82,17h)
plt.subplot(1,3,2)
sce_index = 82
t_index = 17
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,1)
plt.plot(diam, gamma_1d_dist[0,sce_index, t_index, :], label='PR')
plt.plot(diam, gamma_1d_dist[1,sce_index, t_index, :], label='CA', ls='--')
# plt.ylabel(r'$S_i$')
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 0.9, r'sce:%d T:%d h'% (sce_index, t_index))
plt.legend(loc=1)
plt.grid()

# Below 1:1 Line (29,5h)
plt.subplot(1,3,3)
sce_index = 29
t_index = 5
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,1)
plt.plot(diam, gamma_1d_dist[0,sce_index, t_index, :], label='PR')
plt.plot(diam, gamma_1d_dist[1,sce_index, t_index, :], label='CA', ls='--')
# plt.ylabel(r'$S_i$')
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 0.9, r'sce:%d T:%d h'% (sce_index, t_index))
plt.legend(loc=1)
plt.grid() 

plt.tight_layout()

# plt.savefig('/data/keeling/a/yicenl2/d/images/h2_hgamma/h2.png')
```

# 1D Size Distribution (h2, surf_area_concs)

```python
directory = 'sce_completed/scenarios'
surf_area_dist = np.zeros((2,100,25,180))
mass_dist = np.zeros((2,100,25,180))

for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ('out/urban_plume_00000' in filename and '.nc' in filename):
#             print(filename)
            t_index = int(filename[-13:-11]) - 1
            sce_index = int(filename[-38:-36])
            ncf = scipy.io.netcdf_file(filename)
            diam = ncf.variables["diam"].data* 1e6 # m to Âµm
            surf_area_dist[0,sce_index, t_index, :] = ncf.variables["surf_area_dist_pr"].data*1e6 # m2 to mm2
            surf_area_dist[1,sce_index, t_index, :] = ncf.variables["surf_area_dist_comp"].data*1e6
            mass_dist[0,sce_index, t_index, :] = ncf.variables["mass_dist_pr"].data*1e6 # kg to mg
            mass_dist[1,sce_index, t_index, :] = ncf.variables["mass_dist_comp"].data*1e9
```

```python
plt.figure(figsize=(18,4))

# Above 1:1 Line (74,4h)
plt.subplot(1,3,1)
sce_index = 59
t_index = 14
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,2.1e3)
plt.plot(diam, surf_area_dist[0, sce_index, t_index, :], label='surf_area', lw=1.5, color='#1f77b4')
plt.ylabel(r'Surface Area ($\rm \mu m^2/cm^3$)', color='#1f77b4')
plt.tick_params(axis='y', colors='#1f77b4')
plt.xlim(1e-2,1e3)
plt.ylim(0,500)
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 455, r'sce:%d T:%d h'% (sce_index, t_index))
ax = plt.twinx()
p2 = ax.plot(diam, mass_dist[0, sce_index, t_index, :], label='mass', lw=1.5, color='green')
ax.set_ylim(0,100)
ax.tick_params(axis='y', colors='green')
ax.set_ylabel('Mass (mg)', color='green')
plt.grid()

# On 1:1 Line (82,17h)
plt.subplot(1,3,2)
sce_index = 82
t_index = 17
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,2.1e3)
plt.plot(diam, surf_area_dist[0, sce_index, t_index, :], label='surf_area', lw=1.5, color='#1f77b4')
plt.ylabel(r'Surface Area ($\rm \mu m^2/cm^3$)', color='#1f77b4')
plt.tick_params(axis='y', colors='#1f77b4')
plt.xlim(1e-2,1e3)
plt.ylim(0,100)
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 90, r'sce:%d T:%d h'% (sce_index, t_index))
ax = plt.twinx()
p2 = ax.plot(diam, mass_dist[0, sce_index, t_index, :], label='mass', lw=1.5, color='green')
ax.set_ylim(0,100)
ax.tick_params(axis='y', colors='green')
ax.set_ylabel('Mass (mg)', color='green')
plt.grid()

# Below 1:1 Line (47,18h)
plt.subplot(1,3,3)
sce_index = 47
t_index = 18
plt.xscale('log');plt.xlim(1e-2, 1e3);plt.ylim(0,2.1e3)
plt.plot(diam, surf_area_dist[0, sce_index, t_index, :], label='surf_area', lw=1.5, color='#1f77b4')
plt.ylabel(r'Surface Area ($\rm \mu m^2/cm^3$)', color='#1f77b4')
plt.tick_params(axis='y', colors='#1f77b4')
plt.xlim(1e-2,1e3)
plt.ylim(0,100)
plt.xlabel(r'$D_{\rm dry}$ ($\mu$m)')
plt.text(0.015, 90, r'sce:%d T:%d h'% (sce_index, t_index))
ax = plt.twinx()
p2 = ax.plot(diam, mass_dist[0, sce_index, t_index, :], label='mass', lw=1.5, color='green')
ax.set_ylim(0,100)
ax.tick_params(axis='y', colors='green')
ax.set_ylabel('Mass (mg)', color='green')
plt.grid()

plt.tight_layout()

# plt.savefig('/data/keeling/a/yicenl2/d/images/h2_hgamma/h2.png')
```

# 1D Size Distribution (hgamma = h1/h2)

```python
directory = 'sce_completed/scenarios'
gamma_1d_dist = np.zeros((2,100,25,180))
surf_area_dist = np.zeros((2,100,25,180))
hgamma = np.zeros((2,100,25,180))

for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ('out/urban_plume_00000' in filename and '.nc' in filename):
            t_index = int(filename[-13:-11]) - 1
            sce_index = int(filename[-38:-36])
            ncf = scipy.io.netcdf_file(filename)
            diam = ncf.variables["diam"].data* 1e6 # m to Âµm
            gamma_1d_dist[0,sce_index, t_index,:] = ncf.variables["gamma_size_pr"].data*1e6 # m3 to cm-3
            gamma_1d_dist[1,sce_index, t_index,:] = ncf.variables["gamma_size_comp"].data*1e6
            surf_area_dist[0,sce_index, t_index,:] = ncf.variables["surf_area_dist_pr"].data*1e6 # m3 to cm-3
            surf_area_dist[1,sce_index, t_index,:] = ncf.variables["surf_area_dist_comp"].data*1e6
            hgamma[0,sce_index, t_index,:] = gamma_1d_dist[0,sce_index, t_index,:]/surf_area_dist[0,sce_index, t_index,:]
            hgamma[1,sce_index, t_index,:] = gamma_1d_dist[1,sce_index, t_index,:]/surf_area_dist[1,sce_index, t_index,:]
```
