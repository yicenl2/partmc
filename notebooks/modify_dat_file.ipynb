```python
import sys, os
import matplotlib
import scipy.io, numpy
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import netCDF4
import matplotlib as mpl
import seaborn as sns
```

```python
# gas_emit.dat

f = '/data/keeling/a/yicenl2/d/project_n2o5/code_scenario_generator/cases/case_1_n2o5/dat_files/gas_emit_0.dat'
df = pd.read_csv(f, sep = '\s+', skiprows = 3)
print(df.shape)
print(df)
```

```python
for i in np.arange(31,49,1):
    df.insert(int(i), str((i+1) * 3600), '')
    df[str((i+1) * 3600)] = df.iloc[:, i-24]
print(df)
df.to_csv("gas_emit_2d.dat", sep = "\t", index = False)
```
