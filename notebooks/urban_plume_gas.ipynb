## Example: Gas mixing ratio 

```python
#Read data
import sys, os
import matplotlib
import scipy.io, numpy
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import netCDF4
import matplotlib as mpl
import seaborn as sns
mpl.rcParams['font.size'] = 12

gas_mixing_ratio = np.zeros((49, 77))
time = np.zeros(49)
directory = '/data/keeling/a/yicenl2/c/git/project/partmc/scenarios/1_urban_plume/out_nh_nea_48'
for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ("urban_plume_0001" in filename) and (".nc") in filename:
            i = int(filename[-5:-3])
            ncf = scipy.io.netcdf_file(filename, mmap=False)
            gas_mixing_ratio[i-1,:] = ncf.variables["gas_mixing_ratio"].data
            time[i-1] =  ncf.variables["time"].data/3600
            
#time = np.linspace(0,48,49)
print(time)
#Make plot
p1 = plt.plot(time, gas_mixing_ratio[:,4], label='NO', lw=2.5)
plt.grid()
ax = plt.twinx()
p2 = ax.plot(time, gas_mixing_ratio[:,10], label=r'$\rm O_3$', lw=2.5, color='green')
ax.set_ylim(5,200)
ax.tick_params(axis='y', colors='green')
ax.set_ylabel(r'$\rm O_3$ mixing ratio', color='green')

p3 = p1 + p2
labs = [l.get_label() for l in p3]
plt.legend(p3, labs, loc=2)
#plt.xlim(0,49)
```

```python
#Read data
import sys, os
import matplotlib
import scipy.io, numpy
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import netCDF4
import matplotlib as mpl
import seaborn as sns
mpl.rcParams['font.size'] = 12

bc_conc = np.zeros((4, 121))
# time = np.zeros(121)
directory = '/data/keeling/a/yicenl2/c/git/project/partmc/scenarios/1_urban_plume'
for subdir, dirs, files in os.walk(directory):
    for file in files:
        filename = subdir + os.sep + file
        if ("nea_5d/urban_plume_process" in filename) and (".nc") in filename: 
            case = int(filename[-38:-37])
            ncf = scipy.io.netcdf_file(filename, mmap=False)
            bc_conc[case-1, :] = ncf.variables["tot_bc_mass_conc"].data*1e9
            
time = np.linspace(0,120,121)
print(time)
```
